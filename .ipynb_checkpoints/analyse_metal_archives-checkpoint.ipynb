# What are the predominant genres and themes in metal music and where they are most popular

In this workbook we are working with the data we web scraped from metal-archives.com and stored in the metal-archives.db database. For further information about the data and how it was collected take a look at the README.md. 

First we have to explore and clean the data to finaly get answers to the following questions:
- Which countries are most the metal bands from in total numbers and by population?
- What are the most popular sub-genres?
- What are the most popular lyrical themes?
- Are there differences in the questions above for bands that are independent or signed to a label?
- Are there differences in the questions above for bands from different world regions?

## Explore and clean the data
To work with our data we first have to read the database (table "band_info") into a pandas dataframe.

```python
import sqlite3
import numpy as np
import pandas as pd

conn = sqlite3.connect('metal-archives.db') 
          
sql_query = pd.read_sql_query ('''
                               SELECT * FROM band_info
                               ''', conn)

band_info = pd.DataFrame(sql_query).set_index('band_id')
band_info.info()
```

It shows that there are all of the 12 columns we scraped from the website and a total of 163965 band entries. Besides band_id which are integers all other data is in text format. The columns formed_in, years_active and data_retrieved should be in a numerical or date format, but since we don't need this data for our analysis we just ignore it.

In the next step we have to look for duplicates that could have been created during the scraping process and drop them. Therefore we check the column "url". Each individual url should only be ones in our data. 

```python
band_info["url"].duplicated().value_counts()
```

```python
band_info.drop_duplicates(subset=['url'], keep='first', inplace=True)
band_info["url"].duplicated().value_counts()
```

769 duplicates were droped. 

Now we have to look for missing data. The information at the top show that there are no null/NaN values in any of our columns. To make sure that this is really the case it is necessary to look deeper in the columns values. There could be values like "Unknown" or "N/A".

```python
band_info.apply(pd.value_counts).loc[["N/A","Unknown","None","Missing"],]
```

```python
band_info["status"].value_counts(dropna=False)
```

```python
band_info.loc[band_info["country_of_origin"].isin(["Unknown", "International"]),"country_of_origin"].value_counts()
```

It turns out that there are indeed missing values:

The columns "location", "formed_in", "lyrical_themes" and "Years_active" have values "N/A" in string format. 
- For our analysis we only need the lyrical themes. So we change this values to NaN and drop the other columns completely. We can keep the NaN values in "lyrical_themes", because we don't need this column to answer all of our questions. For the questions we need it though we just set the concerned rows aside.

There are also "Unknown" values in some columns:
- 9 band names are "Unknown", but in this case these are real values. 
- In the "country_of_origin" column we have values like "Unknown" and "International". We want to analize the dependency of location, so we can't handle this values properly. Since only a few values are affected we drop these rows.

### Normalize genre and lyrical themes

As shown in the table above there can be multiple values in the "genre" and "lyrical_themes" columns. To analyse which of these genres/themes are most common, we have to split the values into a normalized table. 

#### Genre

```python
band_info["genre"].value_counts()
```

For genre there are 7268 different values in our data. This huge variety is because of the combination of multible genre (e.g. "Hardcore Punk, Heavy Metal") and a lot of subgenres that are also splitted in more subgenres (e.g. "Melodic Death Metal" which is the fifth most common entry is a subgenre of "Death Metal" which is the second most common entry). 

In order to be able to make meaningful evaluations it is nessecary to limit the number of possible genres. On metal_archives.com itself you can browse bands by genre. In total there are 16 subgenres to choose from: "Black", "Death", "Doom/Stoner/Sludge", "Electronic/Industrial", "Experimental/Avant-garde", "Folk/Viking/Pagan", "Gothic", "Grindcore", "Groove", "Heavy" "Metalcore/Deathcore", "Power", "Progressive", "Speed", "Symphonic" and "Thrash" . We will stick to this division.

Besides that we have to make it possible that bands can still belong to several of the limited genres. To implement that we create a new table with columns for each genre. On default we set all values to False.

```python
band_genre = band_info[["genre","country_of_origin","current_label"]].assign(black=False, death=False, doom_stoner_sludge=False, 
            electronic_industrial=False, experimental_avantgarde=False, folk_viking_pagan=False, gothic=False, 
            grindcore=False, groove=False, heavy=False, metalcore_deathcore=False, power=False, progressive=False, 
            speed=False, symphonic=False, thrash=False)
band_genre.head()
```

In the next step we loop through the genre column, split the containing strings and compare the words to our new 16 subgenres. If there is a match the value of the specific column will be changed to True. 

```python
import re

def normalize_genre(genre_string, band_id):
    word_list = re.split("[,-;/\s\n]",genre_string)
    
    for word in word_list:
        if word.lower() == "black":
            band_genre.loc[band_id,"black"] = True
        elif word.lower() == "death":
            band_genre.loc[band_id,"death"] = True
        elif word.lower() in ["doom","stoner","sludge"]:
            band_genre.loc[band_id,"doom_stoner_sludge"] = True
        elif word.lower() in ["electronic","industrial"]:
            band_genre.loc[band_id,"electronic_industrial"] = True
        elif word.lower() in ["experimental","avant"]:
            band_genre.loc[band_id,"experimental_avantgarde"] = True
        elif word.lower() in ["folk","viking","pagan"]:
            band_genre.loc[band_id,"folk_viking_pagan"] = True
        elif word.lower() == "gothic":
            band_genre.loc[band_id,"gothic"] = True
        elif word.lower() == "grindcore":
            band_genre.loc[band_id,"grindcore"] = True
        elif word.lower() == "groove":
            band_genre.loc[band_id,"groove"] = True
        elif word.lower() == "heavy":
            band_genre.loc[band_id,"heavy"] = True
        elif word.lower() in ["metalcore","deathcore"]:
            band_genre.loc[band_id,"metalcore_deathcore"] = True
        elif word.lower() == "power":
            band_genre.loc[band_id,"power"] = True
        elif word.lower() == "progressive":
            band_genre.loc[band_id,"progressive"] = True
        elif word.lower() == "speed":
            band_genre.loc[band_id,"speed"] = True
        elif word.lower() == "symphonic":
            band_genre.loc[band_id,"symphonic"] = True
        elif word.lower() == "thrash":
            band_genre.loc[band_id,"thrash"] = True         
```

```python
band_genre.apply(lambda row: normalize_genre(row["genre"], row.name), axis = 1)
band_genre.head()
```

```python
band_genre.iloc[:,3:].sum(axis = 1).value_counts()
```

As aspected most bands are only assigned to one or two genres, but some have up to 5 genres. 633 band are not assigned to any genre. We will take a closer look to this cases.

```python
band_genre["genre"].loc[band_genre.iloc[:,3:].sum(axis = 1) == 0,].value_counts().head(10)
```

Most entries are "Goregrind" which is a subgenre of the already existing genre "Grindcore". There are also a lot of values having the prefix "Post". "Post-Metal" as a genre is a bit controversial. On one hand it is used as a subgenre of "Doom-Metal", but also as prefix to other genres for bands that are using musical influences beyond the typical metal style. In the second case we are not talking about a definable genre, so we will not consider these entries. To consider the first case we will look for the exact string "Post-Metal" to add it to the "Doom" genre.

```python
def normalize_genre_added(genre_string, band_id):
    word_list = re.split("[,;/\s\n]",genre_string)
    
    for word in word_list:
        if word.lower() == "goregrind":
            band_genre.loc[band_id,"grindcore"] = True
        elif word.lower() == "post-metal":
            band_genre.loc[band_id,"doom_stoner_sludge"] = True
```

```python
band_genre.apply(lambda row: normalize_genre_added(row["genre"],row.name), axis = 1)

band_genre.iloc[:,3:].sum(axis = 1).value_counts()
```

There are only 231 bands left with no assignment to a genre. That are only 0,003% of all entries, so we just delete this rows.

```python
band_genre = band_genre.drop(band_genre["genre"].loc[band_genre.iloc[:,3:].sum(axis = 1) == 0,].index)

band_genre.iloc[:,3:].sum(axis = 0).sort_values(ascending=False)
```

```python
band_genre.shape
```

Finaly we managed to assign most bands to at least one superior metal genre. We will keep working with the cleaned dataframe (band_genre) later on. It still contains 88401 of the 88631 active bands from our cleaned dataframe.

#### Lyrical theme

```python
band_info["lyrical_themes"].value_counts()
```

As aspected there is a very large variety in lyrical themes. There are 34067 different combinations of themes in our data. As with the genres we have to split the values and create a normalited table. Different to before we can't work with a list of superordinate terms, but have to include every theme.

First we create a new dataframe only containing non-null values in the "lyrical_themes" column.

```python
band_themes = band_info.loc[band_info["lyrical_themes"].notnull(),["lyrical_themes","country_of_origin","current_label"]]#.reset_index(drop=True)
band_themes.info()
```

45166 entries are left for our analysis. 
 
Now we can generate a list of all the unique themes there are. We split the string by comma and space. So we limit the possible themes from word groups to single words. Most entries are single words nevertheless, but so we also can take phrases/sentences in account. In addition we remove a list of very common words like "the" or "of".

```python
band_themes["split_themes"] = band_themes["lyrical_themes"].apply(lambda x: re.split(', | ', x))
band_themes.head()
```

```python
themes_vocabulary = []
for themes in band_themes["split_themes"]:
    for theme in themes:
        themes_vocabulary.append(theme.lower())

# common "filler" words to remove
common_words = ["of", "and", "the", "in", "from", "to", "a", "that", "it", "not", "on", "with", "as", "'n'", "(early);", "(later)"]

themes_vocabulary = set(themes_vocabulary) - set(common_words)
len(themes_vocabulary)
```

Our vocabulary of themes contains 6914 unique words. Each of this words will get its own column in the dataframe. If the theme is applicable to the band it gets the boolean value True. To do this we first create a dictionary with default value False. After looping through the bands data and assigning True values we convert the dictionary to a dataframe and join it with the band_themes dataframe.

```python
assign_theme = {unique_theme: [False] * len(band_themes["split_themes"]) for unique_theme in themes_vocabulary}

for index, themes in enumerate(band_themes["split_themes"]):
    for theme in themes:
        if theme.lower() in common_words: #skip common words
            continue
        assign_theme[theme.lower()][index] = True
```

```python
assign_theme = pd.DataFrame(assign_theme, index=band_themes.index)
assign_theme.head()
```

```python
band_themes = pd.concat([band_themes, assign_theme], axis=1)
band_themes.head()
```

```python
band_themes.iloc[:,4:].sum(axis = 0).sort_values(ascending=False).head(10)
```

```python
band_themes.shape
```

Finaly we have a normalized dataframe which allows to analyse single lyrical themes. We will keep working with the cleaned dataframe (band_themes) later on. It still contains 45166 of the 88631 active bands from our cleaned dataframe.

## Analysing the data

### Which countries are the most metal bands from in total numbers and by population?

To answer this question we first create a new dataframe which contains every country and its total number of appearance. We achive this by converting the output of pd.value_counts() into a dataframe and rename the column.

```python
band_location = pd.DataFrame(band_info["country_of_origin"].value_counts()).rename(columns={"country_of_origin": "total_number"})
```

```python
band_location.head(20)
```

It can be seen immediately that the most bands come from the United States by far. To answer if this is also the case when related to the countries population we have to merge the worlds_popultion dataframe. We need the columns "value" (renamed to "population") and "region".

```python
band_location = band_location.merge(worlds_popultion[["value","region"]], how="left", left_index=True, right_index=True).rename(columns={"value": "population"})
```

```python
band_location.info()
band_location.head()
```

To use the new population data we have to convert it to a numerical data type. We have to remove the commas for that.

```python
band_location["population"] = pd.to_numeric(band_location["population"].str.replace(",",""))
band_location.head()
```

Now we can calculate the number of bands related to their countries population. We calculate this key figure per 100k inhabitants, as this gives us results in the one to two-digit range.

```python
band_location["total_per_100k"] = band_location["total_number"] / band_location["population"] * 100000
band_location.head()
```

It would be very descriptive to represent this key figure on a geographic heat map. There is a python library called GeoPandas that makes it easy to plot such a chart using the Matplotlib library. Rendering maps with GeoPandas requires a shapefile which is a dataframe with some graphical data attached. We use a world map from https://www.naturalearthdata.com/downloads/10m-cultural-vectors/10m-admin-0-details/.

### What are the most popular sub-genres?

```python
band_genre_region = band_genre.merge(worlds_popultion["region"], how="left", left_on="country_of_origin", right_index=True)
```

Now we can generate the frequency table we need by using the functions value_counts() and sum(). There will be a column for total number, signed number, indipendent number and for every of the 10 world regions.

```python
regions = band_genre_region["region"].unique()
regions
```

```python
genre_matrix = pd.DataFrame(band_genre_region.iloc[:,3:-1].sum(axis = 0).sort_values(ascending=False)).rename(columns={0:"Total"})
genre_matrix["Total Signed"] = band_genre_region[band_genre_region["current_label"] != "Unsigned/independent"].iloc[:,3:-1].sum(axis = 0)
genre_matrix["Total Unsigned"] = band_genre_region[band_genre_region["current_label"] == "Unsigned/independent"].iloc[:,3:-1].sum(axis = 0)

for region in regions:
    genre_matrix[region] = band_genre_region[band_genre_region["region"] == region].iloc[:,3:-1].sum(axis = 0)

genre_matrix
```

To determine whether a genre is related to status of signing or region, the chi-square test is used. If the resulting p-value is less than 5%, the null hypothesis is rejected. Which would mean that there is a dependency between the variables.

Additionally, we calculate Cramer's V. It returns a value between 0 and 1. If the value returned is greater than 0.5, the relationship is considered very strong. If the value is less than 0.10, the relationship is very weak.

```python
import scipy.stats as stats

chi_stats_genre_label = stats.chi2_contingency(genre_matrix.iloc[:,1:3])
X2 = chi_stats_genre_label[0]
n = np.sum(genre_matrix["Total"])
minDim = min(genre_matrix.iloc[:,1:3].shape)-1
# Cramers V berechnen
V = np.sqrt((X2/n) / minDim)
print("p-value genre/label: ", chi_stats_genre_label[1])
print("cramers v genre/label: ", V)

chi_stats_genre_region = stats.chi2_contingency(genre_matrix.iloc[:,3:])
X2 = chi_stats_genre_region[0]
n = np.sum(genre_matrix["Total"])
minDim = min(genre_matrix.iloc[:,3:].shape)-1
# Cramers V berechnen
V = np.sqrt((X2/n) / minDim)
print("p-value genre/region: ", chi_stats_genre_region[1])
print("cramers v genre/region: ", V)
```

Both p-values are 0, which means that there is a dependency between genre and status of signing and also region. With very small values for Cramer's V, however, this dependency is very weak.

To make the data comparable we convert the frequencies to percentages. In addition we rename column and index names so that they are easier to read on a chart.

We use the Statsmodel library to generate a statsmodels table object that enables the resid_pearson method. This method returns a table of residuals which will reveal any associations present in the data. Positive values indicate more observations than expected and negative values indicate fewer. By dividing this table by the numeber of observations we get the percentage deviations from the expected distribution.
